```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame(data)

print(df.head())
```

```python
# Summary
print(df.describe())

# Create Average column
df["Average"] = (df["Math"] + df["Science"] + df["English"]) / 3

print(df)
```

```python
top_students = df.sort_values(by="Average", ascending=False)

print("Top Students:")
print(top_students[["Name", "Average"]].head())
```

```python
gender_avg = df.groupby("Gender")["Average"].mean()

print("Average Marks by Gender:")
print(gender_avg)
```

```python
# Study Hours vs Marks
plt.scatter(df["Study_Hours"], df["Average"])
plt.xlabel("Study Hours")
plt.ylabel("Average Marks")
plt.title("Study Hours vs Marks")
plt.show()

# Attendance vs Marks
plt.scatter(df["Attendance"], df["Average"])
plt.xlabel("Attendance")
plt.ylabel("Average Marks")
plt.title("Attendance vs Marks")
plt.show()
```
